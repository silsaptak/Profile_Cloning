```python
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/Saptak/Desktop/Thesis_doc/Insta_paper+dataset/user_fake_authentic_2class.csv")
```

```python
display(data.head())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
X = data[['pos', 'flw', 'flg','bl','pic','lin','cl','cz','ni','erl','erc','lt','hc','pr','fo','cs','pi']]  # Features
y = data['class']
```

```python
#random_state=42 ensures that every time you run the code, the split will be the same. This helps maintain consistency in the results when you share the code with others or rerun it multiple times.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#Random forest apply( n_estimators - no of trees)

clf = RandomForestClassifier(n_estimators=120, random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.4f}%")
print("Classification Report:")
print(report)
```
